To create a folder in working directory and then capturing images and saving them into a folder ,ittration for 5 times with 5 different signs

```python
for i in range(1,6):
        os.mkdir(r"/home/dharani/Documents/images for prgmg/opencv/framin{}".format(i))
        if input("entre to proceed:")=="f":
            print("videocapture started")
            vi=cv2.VideoCapture(0)
            print("videocapture ended")
            count=0
            while True:
                    count=count+1
                    su,image=vi.read()
                    print(su)
                    ## it will return 2 things , 1.whether frames are there or not,2.copy of the frame in to img
                    cv2.imshow("video",image)
                    cv2.imwrite(r"/home/dharani/Documents/images for prgmg/opencv/framin{}/{}.jpg".format(i,count),image)
                    if cv2.waitKey(1)==ord("q"):
                                vi.release()
                                cv2.destroyAllWindows()
                                break
            
```

listdir() method in python is used to get the list of all files and directories in the specified directory

```python
from num2words import num2words
list=[]
for j in range(1,6):
    data=pd.DataFrame()
    for i in range(1,60):
            flat=cv2.imread(r"/home/dharani/Documents/images for prgmg/opencv/framin"+str(j)+"/"+str(i)+'.jpg').flatten()
            data=data.append({'vectors':flat,'label':num2words(j)},ignore_index=True)
    list.append(data)
```
